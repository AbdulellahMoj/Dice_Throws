```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sides = 6

my_dice = np.array(range(1,sides+1))

my_dice

```

```python
#roll the dice "random"

np.random.choice(my_dice)
```

```python
# Roll the dice n times
n_rolls = 100

# Save the result of each roll
rolls = np.array([np.random.choice(my_dice) for _ in range(n_rolls)])

rolls
```

```python
# mean of the 100 rolls
mean = rolls.mean()

#variance of the 100 rolls
var = rolls.var()

print(f'The mean of the 100 rolls is {mean}')
print(f'The variance of the 100 rolls is {var}')
```

```python
# display
n_rolls_hist = sns.histplot(rolls, discrete=True)
n_rolls_hist.set(title=f"{n_rolls} rolls")
plt.show()
```

```python
# try a bigger n of rolls like 100,000

n_rolls = 100_000

rolls = np.array([np.random.choice(my_dice) for _ in range(n_rolls)])

print(f"mean of rolls: {np.mean(rolls):.2f}\nvariance of rolls: {np.var(rolls):.2f}")

n_rolls_hist = sns.histplot(rolls, discrete=True)
n_rolls_hist.set(title=f"{n_rolls} rolls")
plt.show()
```
